```python
import pandas as pd

# Initialize empty lists to store data
author_ids = []
texts = []

def readfile(filename):
# Open and read the file line by line
    with open(filename, 'r') as file:
        lines = file.readlines()

        # Initialize flags
        is_author_line = False
        is_body_text = False

        # Iterate through each line in the file
        for line in lines:
            # Extract author id
            if 'author id=' in line:
                author_id = line.split('"')[1]
                author_ids.append(author_id)
                is_author_line = True
            elif is_author_line:
                is_author_line = False

            # Extract text between <body> and </body>
            if '<body>' in line:
                is_body_text = True
                body_text = ''
            elif '</body>' in line:
                is_body_text = False
                text=body_text.strip()
                cleaned_text = text.replace('\n', '')
                texts.append(cleaned_text)
            elif is_body_text:
                body_text += line

    # Create a pandas DataFrame
    df = pd.DataFrame({'AuthorID': author_ids, 'Text': texts})
    return df

# Display the DataFrame

```

```python
def readfile_text(filename):
# Open and read the file line by line
    with open(filename, 'r') as file:
        lines = file.readlines()
        texts=[]
        author_ids=[]
        # Initialize flags
        is_author_line = False
        is_body_text = False

        # Iterate through each line in the file
        for line in lines:

            # Extract text between <body> and </body>
            if '<body>' in line:
                is_body_text = True
                body_text = ''
            elif '</body>' in line:
                is_body_text = False
                text=body_text.strip()
                cleaned_text = text.replace('\n', '')
                texts.append(cleaned_text)
            elif is_body_text:
                body_text += line

    # Create a pandas DataFrame
    df = pd.DataFrame({ 'Text': texts})
    return df

def readfile_Label(filename):
# Open and read the file line by line
    with open(filename, 'r') as file:
        lines = file.readlines()
        texts=[]
        author_ids=[]
        # Initialize flags
        is_author_line = False
        is_body_text = False

        # Iterate through each line in the file
        for line in lines:
            # Extract author id
            if 'author id=' in line:
                author_id = line.split('"')[1]
                author_ids.append(author_id)
                is_author_line = True
            elif is_author_line:
                is_author_line = False
    # Create a pandas DataFrame
    df = pd.DataFrame({ 'AuthorID': author_ids})
    return df
```
